### replacing long unwieldy titles with shorter ones

```python
df.columns = (df.columns
              .str.strip()
              .str.replace(r'How did you hear about County College of Morris\?', 'heard_from', regex=True)
              .str.replace(r'To what extent did the following impact your decision to attend County College of Morris\?', 'impact_ccm', regex=True)
              .str.replace(r'Which course are you enrolled in?', 'course')
              .str.replace(r'Did you receive information about the CCM computing programs from any of the following sources?', 'info_sources', regex=True)
              .str.replace(r'Was a computing major/certificate your first choice, or did you change majors from a different CCM program? If you changed majors, indicate what your first major was.', 'change_major')
              .str.replace(r'On a scale of 1 to 5, with 1 being not at all interested and 5 being extremely interested, how interested are you in taking more computing classes?', 'interest')
              .str.replace(r'Please explain your answer to the question above.  Why or why not would you be interested in taking another computing class?', 'explain')
              .str.replace(r'Race/ethnicity', 'race')
              .str.replace(r'Prior to applying to college, did you participate in any of the following events or activities at the County College of Morris and/or with the Department of Information Technologies, if at all?', 'events', regex=True)
              .str.replace(r'What degree program are you currently enrolled in', 'degree')
              .str.replace(r'What motivated you to seek a computing degree/certificate at CCM?', 'motivated', regex=True)
              .str.replace(r'To what extent did the following activities or experience impact your decision to enroll in an computing course at CCM?', 'impact_comp', regex=True)
              .str.replace('?', '', regex=False)
              .str.replace(r'CCM Web site','website')
              .str.replace(r'Community Event', 'event')
              .str.replace(r'Family member or friend', 'familyfriend')
              .str.replace(r'Current CCM student', 'student')
              .str.replace(r'CCM Alumni', 'alumni')
              .str.replace(r'High School Teacher', 'teacher')
              .str.replace(r'High School Counselor','counselor')
              .str.replace(r'In-app advertisement', 'ad')
              .str.replace(r'Affordable cost', 'cost')
              .str.replace(r'Location/convenience', 'location')
              .str.replace(r'Choice of programs', 'choice')
              .str.replace(r'Online offerings','online')
              .str.replace(r'Family/friend referral', 'referral')
              .str.replace(r'Faculty/staff','faculty')
              .str.replace(r'College reputation', 'reputation')
              .str.replace(r'Computing-related competitions (e.g., Robotics competition, Lego competition, Cybersecurity, Programming)','competitions')
              .str.replace(r'Titans Tuesday (Virtual) Information Session', 'virtual info session')
              .str.replace(r'Transfer credits back to HS degree (ShareTime, Challenger Program)', 'hs_transfer')
              .str.replace(r'Ability to transfer CCM credits to a 4-year school', 'transfer credits from ccm')
             )
```

### making columns lowercase, removing brackets, and replacing spaces with underscores

```python
df.columns = df.columns.str.lower().str.replace(' ', '_').str.replace('[','',regex=False).str.replace(']','',regex=False).str.replace(r'_+', '_', regex=True)
```

```python
df.columns
```

## remove irrevelant features
### I want to be careful, so I'm only removing columns that I'm sure are irrelavent. I'll return later into this project if more come up

```python
#remove timestamp column
df.drop(['timestamp','explain'], axis=1, inplace=True)
```

## make all entries lowercase

```python
df = df.apply(lambda x: x.str.lower() if x.dtype == 'object' else x)
```

## remove redundency

```python
df.drop_duplicates()
```

```python
df['degree'].unique()
```

### I saw a lot of different ways of saying engineering

```python
df['degree'] = df['degree'].replace({
    'mech e': 'engineering',
    'electronical engineer' : 'engineering',
    'mechanical technology' : 'engineering',
    'cyberseucirty' : 'cybersecurity',
    'graphic arts' : 'graphic design',
    'data analytics certificate of achievement' : 'data science',
    'information security certificate of achievement' : 'cybersecurity'
})
#trying to be overly fancy here. In the end I didn't save any time, but I learned how to use lambda
df['degree'] = df['degree'].apply(lambda x: 'engineering' if isinstance(x, str) and 'engineering' in x.lower() else x)
df['degree'] = df['degree'].apply(lambda x: 'business' if isinstance(x, str) and 'business' in x.lower() else x)
df['degree'] = df['degree'].apply(lambda x: 'data science' if isinstance(x, str) and 'data science' in x.lower() else x)
df['degree'] = df['degree'].apply(lambda x: 'information technology' if isinstance(x, str) and 'information technology' in x.lower() else x)
```

## cleaning the change major column

### investigate the change major column

```python
df['change_major'].value_counts()
```

### Removing redundancy

```python

df['change_major'] = df['change_major'].replace({
    'i did not change to any other major.': 'first choice',
    'major didn\'t change' : 'first choice',
    'i have a bachelors in criminal justice and came back to add an it as and cybersecurity certificate': 'cybersecurity',
})
```

### exploring the race column for redundancy

```python
df['race'].value_counts()
```

 ### It looks like the last entry just checked everything, so I'll replace it with 'choose not to reply'

```python
df['race'] = df['race'].replace({
    'hispanic or latino;american indian/native american/alaska native;asian;black/african american;native hawaiian/other pacific islander;white/caucasian;multi-racial;choose not to reply' : 'choose not to reply',
    })
```

### Collapsing all entries that selected multiple races into multi racial

```python
mask = df['race'].str.contains(';')
df.loc[mask, 'race'] = 'multi-racial'
```

```python
df['race'].value_counts()
```

### When I think about it, there is very little practical difference between n/a, no impact, and some impact. If something was truely impactful, the student would have written high impact. Therefore, I'll roll those three responses into no impact.

```python
df[df.filter(like='impact').columns] = df.filter(like='impact').fillna('no impact')
```

```python
df.replace({'no impact': 'no impact',
            'some impact': 'no impact'}, inplace=True)
```

### I'm also droping the "other" columns because they are not useful information

```python

df= df.drop(['impact_comp_other','info_sources_other','heard_from_other'], axis = 1)
```
